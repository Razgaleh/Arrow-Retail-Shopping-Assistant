```python
df = pd.read_csv("./shared/data/Product_Dataset_wImages.csv", encoding="latin1")
df['image'] = "/cdw_images/" + df['ProductName'].str.replace(" ", "_") + ".jpg"
clean_df = df.copy()
clean_df['name'] = clean_df['ProductName'].str.title()
clean_df['url'] = clean_df['image']
clean_df['image'] = clean_df['image']
clean_df['category'] = clean_df['LogisticsTaxonomyClassName']
clean_df['subcategory'] = clean_df['LogisticsTaxonomyGroupName']
clean_df['description'] = clean_df['MarketingText']
clean_df = clean_df[['category', 'subcategory', 'name', 'description','url', 'image']]
clean_df
```

```python
unique_df = clean_df.drop_duplicates()
unique_df['price'] = np.round(np.random.uniform(30, 100, size=len(unique_df)), 2)
unique_df = unique_df.reset_index(drop=True)
unique_df = unique_df[['category', 'subcategory', 'name', 'description','url','price', 'image']]
# unique_df.to_csv("./shared/data/clean_cdw_products.csv", index=False)

```
